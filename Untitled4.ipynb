# Overloading

Overloading merupakan salah satu konsep dalam pemrogramaan berorientasi objek dimana dalam suatu class terdapat method dengan nama yang sama namun tetapi memiliki berbagai fungsi yang berbeda, tergantung dari parameter yang digunakan


```python
class point: 
    def __init__(self,x=0,y=0):
        self.x=x 
        self.y=y 
    #def_str_(self): 
    #return"((0),(1))"format(self.x,self.y)" 
    def __str__(self): 
        return "point object is at("+str(self.x)+","+str(self.y)+")" #nilai balik 
    def __sub__(self, other): 
        x=self.x-other.x 
        y=self.y-other.y 
        return point(x,y) 
    def __add__(self,other): 
        x=self.x+other.x 
        y=self.y+other.y 
        return point(x,y)
p1= point(2,3)
p2= point(-1,2)
print (p1-p2) 
print(p1+p2)
```
